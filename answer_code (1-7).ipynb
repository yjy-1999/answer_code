```python
#question 1

def right_sum(m,n,required_sum):
    vertical_sum = (1 + m) * m // 2
    left = required_sum - vertical_sum
    horizontal_values = [1] * (n - 1)
    left -= sum(horizontal_values) 

    for i in range(n - 2, -1, -1):
        current = max(1, min(left + horizontal_values[i], m))
        left -= current - horizontal_values[i] 
        horizontal_values[i] = current
        if left == 0:
            break

    print(required_sum, end=' ')
    
#----"output"----#
    cursor = 1
    for i in range(n - 1):
        while cursor < horizontal_values[i]:
            print('D', end='')
            cursor += 1
        print('R', end='')

    while cursor < m:
        print('D')
        cursor += 1

result=right_sum(9,9,65)
print('')
right_sum(9,9,72)
print('')
right_sum(9,9,90)
print('')
right_sum(9,9,110)
print(' ')
right_sum(90000,100000,87127231192)
print('')
right_sum(90000,100000,5994891682)
```
